```python
import pandas as pd
from datetime import datetime, timedelta

# Data de início e fim para o período de um ano
start_date = datetime(2022, 1, 1)
end_date = datetime(2023, 1, 1)

# Intervalo de uma semana
interval = timedelta(weeks=2)

# Loop para percorrer o ano em intervalos de uma semana
while start_date <= end_date:
    # Calcular a data de início e fim da semana
    week_start_date = start_date
    week_end_date = start_date + interval - timedelta(seconds=1)

    # Chamar a função search_by_keyword para a semana atual
    df = search_by_keyword(f"{week_start_date.isoformat()}Z", f"{week_end_date.isoformat()}Z")

    # Nome do arquivo CSV
    csv_filename = f'{week_start_date.strftime("%Y-%m-%d")}_to_{week_end_date.strftime("%Y-%m-%d")}.csv'

    # Salvar o DataFrame em um arquivo CSV
    df.to_csv(csv_filename, index=False)

    files.download(csv_filename)
    # Avançar para a próxima semana
    start_date += interval
```
